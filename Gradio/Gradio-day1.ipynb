```python
import gradio as gr
```

```python
# pip install --upgrade gradio
def greet(text):
    print(f"Shout has been called with input {text}")
    return text.upper()

greet("example text")

```

```python
simple_chat = gr.Interface(fn=greet, inputs="textbox", outputs="textbox", flagging_mode="never")
```

```python
simple_chat.launch(inbrowser=True, auth=("achie", "achie1"))
```
